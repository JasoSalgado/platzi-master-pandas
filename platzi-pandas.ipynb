```python
#Importing libraries
import pandas as pd
import matplotlib as m
import seaborn as sns

#Reading a CSV
covid_data = pd.read_csv('covid-19-mexico.csv')
df = pd.DataFrame(covid_data)

df.groupby('State')['Age'].max().plot(kind='bar', legend='Reverse')

```

```python
#Importing libraries
import pandas as pd
import matplotlib as m
import seaborn as sns

#Reading a CSV
covid_data = pd.read_csv('covid-19-mexico.csv')
df = pd.DataFrame(covid_data)

sns.lmplot(x='State', y='Age', data=df, fit_reg = False, legend = True, palette='Paired')
```
